```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn import metrics
```

```python
df=pd.read_csv("Salary_dataset.csv")
print(df)
```

```python
plt.plot(df['Candidate'],df['Salary'])
plt.xlabel("Candidate")
plt.ylabel("Salary")
```

```python
plt.scatter(df['Interview_score'],df['Salary'])
plt.xlabel("Interviewscore")
plt.ylabel("Salary")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['Candidate','Interview_score','Experience']],df['Salary'])
```

```python
print(reg.coef_)
print(reg.intercept_)
yp=reg.predict(df[['Candidate','Interview_score','Experience']])
```

```python
plt.plot(df['Candidate'],yp,color='k')
plt.scatter(df['Candidate'],df['Salary'],color='r')
```

```python
plt.plot(df['Experience'],yp,color='k')
plt.scatter(df['Experience'],df['Salary'],color='r')
```

```python
plt.plot(df['Interview_score'],yp,color='k')
plt.scatter(df['Interview_score'],df['Salary'],color='r')
```
